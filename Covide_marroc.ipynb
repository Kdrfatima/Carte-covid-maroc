```python
# ==========================
# 1Ô∏è‚É£ Import des biblioth√®ques
# ==========================
import kagglehub
import pandas as pd
import os

# ==========================
# 2Ô∏è‚É£ T√©l√©chargement du dataset
# ==========================
path = kagglehub.dataset_download("abdelghanibelgaid/covid19-morocco")
print("‚úÖ Dataset t√©l√©charg√© dans :", path)

# ==========================
# 3Ô∏è‚É£ Chargement des fichiers CSV
# ==========================
df_nat = pd.read_csv(os.path.join(path, "covid19-morocco.csv"))
df_reg = pd.read_csv(os.path.join(path, "covid19-morocco-regions.csv"))
df_info = pd.read_csv(os.path.join(path, "regions-morocco.csv"))

print("üìÇ Fichiers charg√©s avec succ√®s !")
print(f"‚Üí covid19-morocco.csv : {df_nat.shape}")
print(f"‚Üí covid19-morocco-regions.csv : {df_reg.shape}")
print(f"‚Üí regions-morocco.csv : {df_info.shape}")

# ==========================
# 4Ô∏è‚É£ Nettoyage et pr√©paration
# ==========================
# Harmoniser les noms
df_reg.columns = df_reg.columns.str.strip().str.lower()
df_info.columns = df_info.columns.str.strip().str.lower()

# V√©rifier les premi√®res lignes
print("\nüîπ covid19-morocco-regions.csv :")
print(df_reg.head())

print("\nüîπ regions-morocco.csv :")
print(df_info.head())

# ==========================
# 5Ô∏è‚É£ Agr√©gation finale par r√©gion
# ==========================
# Les colonnes de r√©gions dans covid19-morocco-regions.csv
region_codes = ['tth','or','fm','rsk','bk','cs','ms','dt','sm','go','ls','do']

# On garde uniquement la derni√®re ligne (valeurs cumul√©es finales)
last_row = df_reg.iloc[-1]

# Construction du dataframe final
data = []
for code in region_codes:
    region_name = {
        'tth': 'Tanger-T√©touan-Al Hoceima',
        'or': 'Oriental',
        'fm': 'F√®s-Mekn√®s',
        'rsk': 'Rabat-Sal√©-K√©nitra',
        'bk': 'B√©ni Mellal-Kh√©nifra',
        'cs': 'Casablanca-Settat',
        'ms': 'Marrakesh-Safi',
        'dt': 'Dr√¢a-Tafilalet',
        'sm': 'Souss-Massa',
        'go': 'Guelmim-Oued Noun',
        'ls': 'La√¢youne-Sakia El Hamra',
        'do': 'Dakhla-Oued Ed-Dahab'
    }[code]
    data.append({
        "Region": region_name,
        "Code": code.upper(),
        "Confirmed": last_row[code],
        "Deaths": last_row["deaths"],
        "Recovered": last_row["recovered"]
    })

df_final = pd.DataFrame(data)

# Harmonize column name in df_final before merging
df_final.rename(columns={'Region': 'region'}, inplace=True)


# Fusion avec le fichier des infos r√©gionales
df_final = df_final.merge(df_info, on="region", how="left")

# ==========================
# 6Ô∏è‚É£ Aper√ßu du r√©sultat
# ==========================
print("\n‚úÖ Donn√©es finales pr√™tes :")
print(df_final.head())

# ==========================
# 7Ô∏è‚É£ Sauvegarde pour visualisation
# ==========================
df_final.to_json("covid_regions_maroc.json", orient="records", force_ascii=False)
print("\nüìÅ Fichier JSON export√© : covid_regions_maroc.json")
```
